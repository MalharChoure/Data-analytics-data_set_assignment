### creating a plot using matrix

# problem no 1: Using numpy linspace function , create an array x containing 500 floats ranging from -2 to +2 then create a second array y as y = 2x-5

```python
x=np.linspace(-2,2,500)
```

```python
y= (2*x)-5
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y,'r--')
plt.title('2x-5')
plt.xlabel('x')
plt.ylabel('y=2x-5')
plt.grid(True)
```
